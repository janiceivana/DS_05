```python
import pandas as pd
import seaborn as se
import matplotlib.pyplot as plt

```

```python
df = pd.DataFrame(pd.read_csv("sell_price.csv"))
sales_train_evaluation =  pd.DataFrame(pd.read_csv("sales_train_evaluation.csv"))
```

```python
dims = (20, 20)
fig, ax = plt.subplots(figsize=dims)

se.boxplot(ax= ax, data= df, x = "sell_price", y="category" , whis= 200)
plt.savefig("Distribution of price between Product Category")
```

```python
sales_train_evaluation['Total_Sales']=sales_train_evaluation.iloc[:,6:1946].sum(axis=1)
grp_sale = sales_train_evaluation.groupby("dept_id", as_index=False)

grp_sale_agg =  grp_sale.aggregate("sum")
```

```python
dims = (20, 20)
fig2, ax = plt.subplots(figsize=dims)
se.boxplot(ax= ax, data= sales_train_evaluation, x = "Total_Sales", y= "dept_id", whis= 100000 )
plt.savefig("Distribution of Sales between Product Category")

grp_sale.describe()
```

```python
fig3, ax = plt.subplots(figsize=dims)
se.barplot(ax= ax, data= grp_sale_agg, x= "Total_Sales", y= grp_sale_agg["dept_id"] )
plt.savefig("Sales between Product Category")
```
